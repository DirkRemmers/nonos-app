```python
import pandas as pd
import numpy as np
import datetime

def transform_to_day(date_string):
    date = date_string.split('-')
    d_list = []
    for d in date:
        d = int(d.split(' ')[0])
        d_list.append(d)
    ans = datetime.date(d_list[0], d_list[1], d_list[2])
    return(ans.strftime("%A"))
```

```python
month_list = ['April', 'Mei', 'Juni', 'Juli', 'Augustus', 'September', 'Oktober']
file_list = ['./Zettle-Raw-Data-Report-20210401-20210430.xlsx','./Zettle-Raw-Data-Report-20210501-20210531.xlsx','./Zettle-Raw-Data-Report-20210601-20210630.xlsx','./Zettle-Raw-Data-Report-20210701-20210731.xlsx','./Zettle-Raw-Data-Report-20210801-20210831.xlsx','./Zettle-Raw-Data-Report-20210901-20210930.xlsx','./Zettle-Raw-Data-Report-20211001-20211031.xlsx']

df_list = []
for m, f in zip(month_list, file_list):
    month_df = pd.read_excel(f, header=5)
    month_df['Maand'] = m
    totaal = np.sum(month_df['Uiteindelijke prijs (EUR)'])
    month_df['Totaal bedrag maand'] = totaal
    month_df['Dag'] = month_df.apply(lambda row: transform_to_day(str(row['Datum'])), axis = 1)
    for d in month_df['Dag'].unique():
        day_df = month_df[month_df['Dag'] == d]
        day_df['Totaal bedrag dag'] = np.sum(day_df['Uiteindelijke prijs (EUR)'])
        df_list.append(day_df)

df = pd.concat(df_list)
```
